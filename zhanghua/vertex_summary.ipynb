```python
import vertexai
from vertexai.preview.generative_models import GenerativeModel, Part
import json

prompt = """以下是幾則對餐廳的評論:

{}

請總結出至多五個有關這間餐廳的分類或是特色, 以逗號分隔, 並且用中文描述.
"""

def generate(reviews):
    model = GenerativeModel("gemini-pro-vision")
    responses = model.generate_content(prompt.format(reviews),
                                       generation_config = {
                                         "max_output_tokens": 2048,
                                         "temperature": 0.7,
                                         "top_p": 1,
                                         "top_k": 32},
                                       stream=False,)
  
    try:
      result_str = responses.candidates[0].content.parts[0].text
    except:
      result_str = ''
  
    return result_str

# result = generate(""""
# 服務人員很親切，還協助提供給寶寶的食物
# 除此之外套餐沙拉也很有誠意不是隨便灑灑玉米粒+苜蓿芽的沙拉，有水果美生菜胡麻醬，蘑菇濃湯好喝附一片麵包。
# 主餐點青醬蛤蜊義大利麵，蛤蜊煮的胖胖的，是剛開殼就起鍋的，這對蛤蜊愛好者很重要，吃起來很鮮甜，青醬味道也很足夠。


# 很驚艷的一間義大利料理餐廳！
# 在埔頂路一帶，外觀並沒有非常起眼
# 裝潢的氣氛不錯，內用空間很寬敞
# 沒有收服務費，服務態度非常好，隨時看到客人的需求，也不會過度打擾用餐，外場服務人員很漂亮有氣質:smiling face:

# 主餐部分可以選擇升級套餐
# 當天選了100: 飲料、沙拉、濃湯三選二

# 超推「海膽風味醬海鮮燉飯」 +焗烤（+30）
# 裡面海鮮量超多，也都很新鮮
# 海膽這個味道還沒有在其他餐廳看過，很特別
# 當天一直偷吃友人的:face with tears of joy:

# 蒜香鮮魷義大利麵
# 魷魚給很多，味道也是很不錯
# 只是海膽風味燉飯太好了吃了被比下去

# 當月壽星有贈送薯條
# 超大一份誠意十足～～

# 整體而言這間店真的很推
# 服務、環境、餐點的水準都很高
# 當天5:30就進去用餐，期間也一直聽到有外送訂單，生意很好
# 期待下次再去用餐！！


# [2023/02/18 初訪]:full moon::full moon::full moon::last quarter moon::new moon:
# 點黃金蒜片舒肥牛義大利麵$270、海膽風味醬海鮮燉飯$270，舒肥牛很嫩，搭配蒜片口味佳，義大利麵偏軟，味道可但稍嫌單調；燉飯一開始吃覺得飯芯稍硬，但飯體摻節瓜，耐吃整體還不錯。

# 付款方式：僅收現金。
# 發票提供：無，免用統一發票。
# 自助飲品：有提供一壺白開水供取用。
# 對外廁所：有，乾淨整潔有洗手乳及擦手紙，並貼心置有生理用品。


# 裝潢氣氛10分
# 服務態度10分
# 餐點外觀10分
# 食材口味9.5
# 今日第一次內用用餐 蘑菇濃湯讓我驚艷 湯頭濃郁香菇爽脆 主菜點紅咖哩雞腿跟金沙鮮蝦 義大利麵的口味都很特別外面比較不常吃到  比較可惜是主食的雞腿跟蝦都有偏一點點柴  不然也會拿10分
# 未來會經常光顧 非常好的用餐體驗


# 主餐跟前菜都好吃～
# 清炒墨魚麵很新鮮 麵的硬度也很好且有入味 不過可惜醬汁略少乾了一點
# 海膽焗飯好吃～裡面有很多海鮮都很新鮮 熟度也剛好
# 前菜奶油燉雞好吃不會膩
# 當月壽星招待滿滿一盤薯條很讚:kissing face with closed eyes:
# 沙拉比較普通一點點""")
```
