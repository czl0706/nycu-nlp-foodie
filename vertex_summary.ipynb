```python
import vertexai
from vertexai.preview.generative_models import GenerativeModel, Part
import json

prompt = """以下是幾則對餐廳的評論:

{}

請總結出至多五個有關這間餐廳的分類或是特色, 以逗號分隔, 並且用中文描述.
"""

def generate(reviews):
    model = GenerativeModel("gemini-pro-vision")
    responses = model.generate_content(prompt.format(reviews),
                                       generation_config = {
                                         "max_output_tokens": 2048,
                                         "temperature": 0.7,
                                         "top_p": 1,
                                         "top_k": 32},
                                       stream=False,)
  
    try:
      result_str = responses.candidates[0].content.parts[0].text
    except:
      result_str = ''
  
    return result_str
```

```python
# load from json
with open('proc_results.json', 'r') as f:
    proc_results = json.load(f)
    
proc_results
```

```python
for idx, result in enumerate(proc_results):
    x = generate(result['reviews'])
    print(idx, result['name'])
    print(x)
    print()
    result['generated_summary'] = x
```

```python
# save as json file
with open('summary_results.json', 'w', encoding='utf-8') as f:
    json.dump(proc_results, f, ensure_ascii=False, indent=4)
```
